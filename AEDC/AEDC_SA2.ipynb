```python
import pandas as pd
import geopandas as gpd
```

```python
SA2_perth = pd.read_excel(r"../SA2_perth.xlsx")
```

```python
sheets = ["Health", "Social", "Emotional", "Language", "Communication"]
        #    "One or more", "Two or more", "Health 1 Subdomain", "Health 2 Subdomain","Health 3 Subdomain" ]
```

```python
def get_column(sheet):
    s = "children_with_valid_scores"
    years = ["2009", "2012", "2015","2018", "2021"]
    attri = "AEDC_"+ sheet +"_"
    out = ["SA2_Code", 
            "Name"]
    for i in years:
        out.append(attri+s+"_"+i)
    other= ["number_on_track",
    "percentage_on_track",
    "number_at_risk", 
    "percentage_at_risk", 
    "number_vulnerable", "percentage vulnerable"]
    years = ["2009", "2012", "2015","2018", "2021"]
    for i in years:
        for j in other:
            out.append(attri+j+"_"+i)
    return out

def get_AEDC(sheet):
    names = get_column(sheet)
    data = pd.read_excel(r"SA2 2009-21 (3).xlsx", sheet_name="SA2 " + sheet)
    data.dropna(axis=1, inplace=True, how="all")
    data = data.iloc[5:]
    data.columns = names
    data.SA2_Code = data.SA2_Code.astype(str)
    data = data[data.SA2_Code.str.len() != 5 ] 
    data = data[(data.SA2_Code.str.contains("/", regex=False) | data.SA2_Code.str.isnumeric())]
    return data
```

```python
master_data = {}
for i in sheets:
    master_data[i] = get_AEDC(i)

final_data =   pd.concat(master_data.values(), axis = 1)
final_data = final_data.loc[:,~final_data.columns.duplicated()].copy()
```

```python
sheets_2 = ["One or more", "Two or more", "Health 1 Subdomain", "Health 2 Subdomain","Health 3 Subdomain" , "On track all"]
def get_column_2(sheet):

    if sheet == "Health 1 Subdomain":
        sheet = "physical_readiness_for_school"
    if sheet == "Health 2 Subdomain":
        sheet = "physical_independence_over_time"
    if sheet == "Health 3 Subdomain":
        sheet = "gross_motor_skills_over_time" 
    s = "children_with_valid_scores"
    years = ["2009", "2012", "2015","2018", "2021"]
    attri = "AEDC_"+ sheet +"_"
    out = ["SA2_Code", 
            "Name"]
    for i in years:
        out.append(attri+s+"_"+i)
    
    other = []
    if sheet=="On track all":
        other = ["number_on_track_for_all", "percentage_on_track_for_all"]
    else:
        other= ["number_vulnerable", "percentage vulnerable"]
    years = ["2009", "2012", "2015","2018", "2021"]
    for i in years:
        for j in other:
            out.append(attri+j+"_"+i)
    return out
def get_AEDC_2(sheet):
    data = pd.read_excel(r"SA2 2009-21 (3).xlsx", sheet_name="SA2 " + sheet)
    data.dropna(axis=1, inplace=True, how="all")
    data = data.iloc[5:]
    names = get_column_2(sheet)
    data.columns = names
    data.SA2_Code = data.SA2_Code.astype(str)
    data = data[data.SA2_Code.str.len() != 5 ] 
    data = data[(data.SA2_Code.str.contains("/", regex=False) | data.SA2_Code.str.isnumeric())]
    return data
```

```python
master_data_2 = {}
for i in sheets_2:
    master_data_2[i] = get_AEDC_2(i)

final_data_2 =   pd.concat(master_data_2.values(), axis = 1)
```

```python
final_data =   pd.concat([final_data, final_data_2], axis = 1)
final_data = final_data.loc[:,~final_data.columns.duplicated()].copy()
```

```python
sa2code = SA2_perth.SA2_CODE21.astype(str)
merged = pd.merge(left=sa2code, right=final_data, left_on = "SA2_CODE21", right_on="SA2_Code", how="inner")
```

```python
merged.to_csv(r"AEDC_SA2.csv")
```
